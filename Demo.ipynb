### Scenario 2: Access Denied - Face Not Recognized

```python
# üö´ SCENARIO 2: Access Denied - Face Not Recognized
# Blurred/unrecognizable face = Access denied at first step

try:
    import cv2
    
    print("üö´ SCENARIO 2: Access Denied - Face Not Recognized")
    print("=" * 50)
    
    # Create a blurred face to simulate unauthorized user
    face_image_path = os.path.join(IMAGE_DIR, 'member1', 'neutral.jpeg')
    voice_sample_path = os.path.join(AUDIO_DIR, 'member2', 'confirm_1.wav')
    
    # Create blurred face for unauthorized attempt
    img = cv2.imread(face_image_path)
    if img is not None:
        blurred_img = cv2.GaussianBlur(img, (99, 99), 30)
        temp_path = 'temp_unauthorized_face.jpg'
        cv2.imwrite(temp_path, blurred_img)
        
        product_access_system(temp_path, voice_sample_path)
        
        # Clean up temp file
        if os.path.exists(temp_path):
            os.remove(temp_path)
    
except NameError:
    print("‚ùå Please run the INITIALIZATION cell at the top of the notebook first!")
```

### Scenario 3: Prediction Blocked - Voice Verification Failed

```python
# ‚ö†Ô∏è SCENARIO 3: Voice Verification Failed
# Face recognition succeeds + Voice verification fails = Prediction blocked

try:
    print("‚ö†Ô∏è SCENARIO 3: Voice Verification Failed")
    print("=" * 50)
    
    # Use face from member2 but voice from member1 (identity mismatch)
    face_image_path = os.path.join(IMAGE_DIR, 'member2', 'neutral.jpeg')
    voice_sample_path = os.path.join(AUDIO_DIR, 'member1', 'confirm_1.wav')
    
    product_access_system(face_image_path, voice_sample_path)
    
except NameError:
    print("‚ùå Please run the INITIALIZATION cell at the top of the notebook first!")
```
