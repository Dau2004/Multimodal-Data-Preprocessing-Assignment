# Product Prediction Model Access System

This notebook demonstrates the exact flow specified:

1. **User attempts to access the product prediction model**
2. **If face is recognized, they proceed to attempt to run a prediction**
3. **The prediction must be confirmed through voice verification**
4. **The system determines if the prediction is allowed**

The system uses pre-trained models to determine whether:
- The face matches a known user (face recognition)
- The voice is an approved voice sample (voiceprint verification)
- The prediction is allowed

### Scenario 2: Access Denied - Face Not Recognized

```python
# Import required functions if not already imported
import cv2
import os

try:
    # Check if variables are already defined
    IMAGE_DIR, AUDIO_DIR, product_access_system
    print("‚úÖ Variables already imported")
except NameError:
    print("üì• Importing required functions...")
    from script import IMAGE_DIR, AUDIO_DIR, product_access_system

# Access denied due to unrecognized face
face_image_path = os.path.join(IMAGE_DIR, 'member1', 'neutral.jpeg')
voice_sample_path = os.path.join(AUDIO_DIR, 'member2', 'confirm_1.wav')

print("üö´ SCENARIO 2: Access Denied - Face Not Recognized")
print("=" * 50)

# Create blurred face for unauthorized attempt
img = cv2.imread(face_image_path)
if img is not None:
    blurred_img = cv2.GaussianBlur(img, (99, 99), 30)
    temp_path = 'temp_unauthorized_face.jpg'
    cv2.imwrite(temp_path, blurred_img)
    
    product_access_system(temp_path, voice_sample_path)
    
    if os.path.exists(temp_path):
        os.remove(temp_path)
```
